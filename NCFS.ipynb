Radar Graphs for Independent Variables

```python
categories = ['Yes', 'No', 'Do not Know', 'Prefer Not to Say', 'NA' ]
categories = [*categories, categories[0]]

High_School = [3345, 2229, 169, 22, 0]
College = [3183, 1494, 74, 14, 1000]
Employer = [2120, 3471, 146, 28, 0]
Military = [423, 534, 29, 11,4768]

High_School = [*High_School, High_School[0]]
College = [*College, College[0]]
Employer = [*Employer, Employer[0]]
Military = [*Military, Military[0]]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(High_School))

plt.figure(figsize=(12, 12))
plt.subplot(polar=True)
plt.plot(label_loc, High_School, label='High School')
plt.plot(label_loc, College, label='College')
plt.plot(label_loc, Employer, label='Employer')
plt.plot(label_loc, Military, label='Military')

plt.title('Financial Education Comparison', size=20)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend()
plt.show()
```
